```python
crime_group = np.sum(crime.groupby('범죄대분류').sum(), axis=1) 

crime_group = pd.DataFrame(crime_group).reset_index()

crime_group.rename(columns = {0:'total'}, inplace = True) 

crime_group.sort_values(by='total', ascending=False, inplace = True)

crime_group
```

```python
crime_list = crime_group['범죄대분류'].tolist()
crime_list
```

```python
plt.figure(figsize = (10, 5)) 
plt.rc('font', family='Malgun Gothic') 
plt.title('범죄대분류 상위 5개 범죄', fontsize=20) 

sns.set_style('darkgrid') 
sns.barplot(data=crime_group, x= '범죄대분류', y = 'total', order = crime_list[:5])

plt.ylabel('총 빈도수', fontsize = 15)
plt.xlabel('범죄대분류', fontsize = 15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.show()
```
